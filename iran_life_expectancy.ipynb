# Visualizing

```python
fig, axes = plt.subplots(nrows=3, figsize=(10, 15))


axes[0].plot(df['Year'], df['Female Life Expectancy'], label='Female Life Expectancy', color='midnightblue', marker = '_')
axes[0].set_title('Female Life Expectancy Over Years')
axes[0].set_xlabel('Year')
axes[0].legend()

# Plot Male Life Expectancy
axes[1].plot(df['Year'], df['Male Life Expectancy'], label='Male Life Expectancy', color='indigo')
axes[1].set_title('Male Life Expectancy Over Years')
axes[1].set_xlabel('Year')
axes[1].legend()

# Plot Life Expectancy Gap
axes[2].plot(df['Year'], df['Life Expectancy Gap'], label='Gap', color='purple')
axes[2].set_title('Life Expectancy Gap Over Years')
axes[2].set_xlabel('Year')
axes[2].legend()

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

# Predicting Female life expectancy

```python
from sklearn.model_selection import train_test_split
x = df[['Population', 'Year']]
y = df['Female Life Expectancy']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state =4)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train , y_train)

y_hat = reg.predict(x_test)
print('y test: ', y_test)
print ('y predicted: ',y_hat)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_hat, y_test)
r2_percentage = r2 * 100
print('The accuracy of your model is: ', r2_percentage)
```
