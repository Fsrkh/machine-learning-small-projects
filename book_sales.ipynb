**Loading data**

```python
import pandas as pd

file = '/content/Books_Data_Clean.csv'

dataset = pd.read_csv(file)
```

```python
print(dataset.shape)
```

```python
dataset.head(10)
```

```python
print(dataset.describe())
```

```python
print(dataset.columns)
```

```python
columns_to_keep = ['Book Name', 'Author', 'Book_average_rating', 'units sold']
dataset = dataset[columns_to_keep]
```

**Visualising Data**

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(figsize=(10, 6))

# Assuming `dataset` is a DataFrame
dataset.plot(kind='box', subplots=True, sharex=False, sharey=False, ax=axes)

plt.show()
```

```python
dataset.hist()
plt.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(dataset)
plt.show()
```
